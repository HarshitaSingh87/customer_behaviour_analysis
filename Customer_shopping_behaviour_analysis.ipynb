```python
#import pandas
import pandas as pd
#load dataset
df = pd.read_csv('customer_shopping_behavior.csv')
#gives first few rows of dataset
df.head()
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(' ','_')
df= df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age_group
labels=['Young_adult','Adult','Middle_aged','Senior']
df['age_group']= pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
if 'Age_Group' in df.columns:
    df = df.drop('Age_Group', axis=1)
df.columns
```

```python
df['purchase_frequency_days']=df['frequency_of_purchases'].str.lower()

# create column purchase_frequency_days
frequency_mapping={
'Fortnightly':14,
'Weekly': 7,
'Monthly': 30,
'Quarterly': 90,
'Bi-weekly': 14,
'Annually': 365,
'Every 3 months': 90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
df=df.drop('promo_code_used',axis=1)
```

```python
if 'Age_Group' in df.columns:
    df = df.drop('Age_Group', axis=1)
df.columns
```

```python
# now we will connect this to postgresql
''' psycopg2-binary = This is the PostgreSQL driver (connector).
It allows Python to talk to PostgreSQL & SQLALchemy = This is a Python ORM / database toolkit.
It simplifies connecting and sending data to SQL databases.'''
!pip install psycopg2-binary sqlalchemy
```

```python
from sqlalchemy import create_engine

#step 1: connect to postgresql
username= "postgres"      # deafult user
password= "Password"      #password set during installation
host= "localhost"
port= "5432"              #default postgre port
database= "customer_behaviour"  #the database created in pgadmin4

engine= create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

#step 2: load data into PostgreSQL
table_name="customer"  #choose any table name
df.to_sql(table_name,engine,if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name} in database '{database}'.")
  
```
